## Exploratory Data Analysis (EDA) Findings
### Executive Summary

The exploratory data analysis identified several factors strongly associated with customer churn. These insights reveal the characteristics of customers most likely to leave and highlight opportunities for targeted retention strategies.

 ### 1.Contract Type is the Strongest Indicator of Churn
Finding

Customers on month-to-month contracts exhibit the highest churn rate, while customers with one-year and two-year contracts are significantly more likely to remain with the company.

Business Impact

Short-term contracts provide customers with greater flexibility to switch providers, increasing customer turnover.

Recommendation
Encourage customers to upgrade to annual or two-year contracts through discounts or loyalty incentives.
Introduce retention offers before month-to-month customers reach renewal dates.

### 2. Customer Tenure is Negatively Related to Churn
Finding

Customers with shorter tenure are substantially more likely to churn than long-term customers.

Business Impact

The highest-risk period occurs during the early months of the customer lifecycle.

Recommendation
Improve onboarding experiences.
Launch engagement campaigns for new customers during their first year.
### 3. Technical Support Reduces Customer Churn
Finding

Customers without Tech Support subscriptions experience noticeably higher churn rates.

Business Impact

Access to technical support appears to improve customer satisfaction and retention.

Recommendation
Promote technical support packages.
Offer free technical support during the first few months of service.

 ### 4. Payment Method Influences Customer Retention
Finding

Customers using Electronic Check payments have a higher churn rate than customers using automatic payment methods.

Business Impact

Manual payment methods may indicate lower customer commitment or create payment friction.

Recommendation
Encourage customers to switch to automatic payment options.
Provide incentives for automatic billing enrollment.

 ### 5. Monthly Charges Influence Churn
Finding

Customers with higher monthly charges tend to churn more frequently.

Business Impact

Price-sensitive customers may seek more affordable alternatives from competitors.

Recommendation
Review pricing strategies.
Introduce personalized discounts for high-value customers at risk of leaving.
 ### 5. Long-Term Customers Generate Greater Customer Lifetime Value (CLTV)
Finding

Customers with longer tenure generally have higher Customer Lifetime Value (CLTV) and lower churn rates.

Business Impact

Retaining existing customers generates greater long-term profitability than acquiring new customers.

Recommendation
Prioritize retention initiatives for high-CLTV customers.
Develop loyalty programs that reward long-term subscribers.
### 7. Competitor Offers are a Major Cause of Customer Churn
Finding

Analysis of Churn Reason indicates that many customers leave because competitors provide better pricing, promotions, or services.

Business Impact

Competitive pressure is a significant driver of customer attrition.

Recommendation
Monitor competitor pricing and promotional campaigns.
Develop competitive retention offers for high-risk customers.
 ### 8. Significant Revenue is at Risk from Customer Churn
Finding

Customers predicted to churn account for a substantial portion of recurring monthly revenue.

Business Impact

Reducing churn would directly improve recurring revenue and overall business profitability.

Recommendation
Identify high-value customers with elevated churn risk.
Implement proactive retention campaigns before contract cancellation.
### 9. Service Adoption Influences Customer Retention
. Finding

Customers subscribed to additional services such as Online Security, Device Protection, and Tech Support generally exhibit lower churn rates than customers without these services.

Business Impact

Customers using multiple services are more engaged with the company's offerings and are less likely to switch providers.

Recommendation
Promote bundled service packages.
Recommend complementary services to existing customers.
 ### 10. Customer Segmentation Reveals High-Risk Groups
Finding

Combining variables such as contract type, tenure, monthly charges, and service subscriptions identifies customer segments with substantially higher churn risk.

Business Impact

Rather than applying broad retention strategies, the company can focus its resources on the customers most likely to leave.

Recommendation
Develop targeted retention campaigns for high-risk customer segments.
Use predictive models to identify these customers before churn occurs.
Overall Business Conclusion

The analysis indicates that customer churn is primarily influenced by contract type, customer tenure, pricing, payment method, service adoption, and competitive pressure. Customers on month-to-month contracts, with shorter tenure, higher monthly charges, and limited service subscriptions are significantly more likely to churn.

These findings provide a strong foundation for the next phase of the project—feature engineering and predictive modeling—where machine learning models will use these patterns to identify customers at high risk of churning and support proactive retention strategies.
